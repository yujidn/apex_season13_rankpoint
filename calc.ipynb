```python
import pandas as pd
```

```python
rps = [0,0,0,0,0,0,0,5,5,5,10,10,20,20,30,45,55,70,95,125]
kps = [1,1,1,1,1,1,1,5,5,5,10,10,12,12,14,16,18,20,23,25]
```

```python
def kill_point(rank_num:int, kill_assist_num:int) -> float:
    kp = kps[20-rank_num]
    if kill_assist_num < 4:
        point = kp * kill_assist_num
    elif kill_assist_num < 6:
        point = kp * 3 + kp * (5 - kill_assist_num) * 0.8
    else:
        point = kp * 3 + kp * 2 * 0.8 + kp * 1 * 0.5 + kp * (kill_assist_num - 6) * 0.2

    return point


def team_point(rank_num:int, team_kill_num:int) -> float:
    kp = kps[20-rank_num]
    if team_kill_num == 0:
        point = 0
    elif team_kill_num == 1:
        point = kp * 0.4
    elif team_kill_num < 4:
        point = kp * 0.4 + kp * (team_kill_num - 1) * 0.1
    else:
        point = kp * 0.4 + kp * 2 * 0.1 + kp * (team_kill_num - 3) * 0.08
        
    return point
```

```python
[(kill_point(1, n), team_point(1, n)) for n in range(1, 20)]
```

```python
col = ["team kill \\ kill assist"] + [i for i in range(1,21)]
```

```python
# rankは1st ~ 20th
for r in range(1,21):
    # killは0kill ~ 20kill
    table = []
    for k in range(0,21):
        row = {"kill assist \\ team kill": k}
        for t in range(0,11):
            kp = kill_point(r, k)
            tp = team_point(r, t)
            rp = rps[20 - r]
            row[str(t)] = int(kp + tp + rp)
        table.append(row)
        
        
    if r == 1:
        pd.DataFrame(table).to_csv('1st.csv')
    elif r == 2:
        pd.DataFrame(table).to_csv('2nd.csv')
    elif r == 3:
        pd.DataFrame(table).to_csv('3rd.csv')
    else:
        pd.DataFrame(table).to_csv(str(r)+'th.csv')
```
